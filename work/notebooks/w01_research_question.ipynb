### Skills Practice: Filtering and Metric Calculation
Following the `skills/README.md` guidelines, we use streaming-friendly methods to explore the data without loading the entire warehouse.

```python
# Example: Calculate CTR for the sample to identify low-performing content
df_sample['ctr'] = (df_sample['gsc_clicks'] / df_sample['gsc_impressions']).fillna(0)

# Filter for content with high impressions but low clicks (potential for refresh)
needs_refresh = df_sample[(df_sample['gsc_impressions'] > 10) & (df_sample['ctr'] < 0.02)]

print(f"Found {len(needs_refresh)} candidates for refresh in this sample slice.")
display(needs_refresh[['content_hash_id', 'gsc_impressions', 'gsc_clicks', 'ctr']])
```

```python
# Example: Skill - Aggregating streaming data manually
# Since we can't .groupby() the whole stream easily, we iterate:
total_impressions = 0
limit = 100

for i, row in enumerate(ds.take(limit)):
    total_impressions += row['gsc_impressions']

print(f"Total impressions across first {limit} rows: {total_impressions}")
```
